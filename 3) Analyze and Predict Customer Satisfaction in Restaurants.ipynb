```python
duplicates = df[df.duplicated()]
# Display duplicate rows
print(duplicates)
```

```python
print(df.tail().hist())
```

```python
sns.lineplot(x='PreferredCuisine',y='MealType',data=df)
```

```python
# Plot distribution of satisfaction scores
plt.figure(figsize=(10, 8))
sns.histplot(df['HighSatisfaction'], bins=20, kde=True)
plt.title('Distribution of Customer Satisfaction Scores')
plt.xlabel('Satisfaction Score')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(x='MealType', y='HighSatisfaction', data=df)
plt.title('Satisfaction Scores by Meal Type')
plt.xlabel('Meal Type')
plt.ylabel('Satisfaction Score')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='FoodRating', y='CustomerID', data=df)
plt.title('FoodRating')
plt.show()
```

```python
sns.boxplot(x='MealType',y='WaitTime',data=df)
```

```python
# Plot pie chart for gender distribution
# Count occurrences of each gender
gender_counts = df['Gender'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels= gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Genders')
plt.show()
```

```python
sns.violinplot(x='TimeOfVisit',y='Age',data=df)
```
