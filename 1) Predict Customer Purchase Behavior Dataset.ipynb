```python
# Step-4 : We will now analyze the data
#Getting histogram based on the first 10 values and last 10 values
print(df.head().hist())
print(df.tail().hist())
```

```python
plt.figure(figsize=(10, 8))
sns.histplot(df['NumberOfPurchases'], bins=30, kde=True)
plt.title('Distribution of NumberOfPurchases')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(x='ProductCategory', y='NumberOfPurchases', data=df)
plt.title('NumberOfPurchases by Product Category')
plt.show()
```

```python
# draw lineplot 
sns.lineplot(x="Age", y="AnnualIncome", data=df)
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
#boxplot
sns.boxplot(x='AnnualIncome',y='NumberOfPurchases',data=df)
```

```python
#  Visualizing the data
#Pair plots
subset_data = df.head()
# Create a pair plot for the first 10 rows
sns.pairplot(subset_data)
plt.suptitle('Pair Plot of First 5 Rows', y=1)
plt.show()
```
