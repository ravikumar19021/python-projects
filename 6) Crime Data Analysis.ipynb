```python
duplicates = df[df.duplicated()]
# Display duplicate rows
print(duplicates)
```

```python
#Since we got duplicates, we must remove them 
data=df.drop_duplicates(inplace=True)
print(data)
```

```python
# Count of each crime type
top_n = 15
top_crimes = df['Crm Cd Desc'].value_counts().nlargest(top_n)
plt.figure(figsize=(15, 10))
sns.barplot(x=top_crimes.values, y=top_crimes.index)
plt.title(f'Top {top_n} Most Common Crime Types', fontsize=16)
plt.xlabel('Number of Incidents', fontsize=14)
plt.ylabel('Crime Type', fontsize=14)
plt.show()
```

```python
# draw lineplot 
sns.lineplot(x="Vict Age", y="Weapon Desc", data=df)
```

```python
sns.barplot(x='Vict Sex',y='Vict Age',data=df)
```

```python
plt.figure(figsize=(15, 12))
sns.histplot(df['Vict Age'].dropna(), bins=30, kde=True)
plt.title('Distribution of Victim Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.countplot(x='Vict Sex', data=df)
plt.title('Gender Distribution of Victims')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(y='AREA NAME', data=df, order=df['AREA NAME'].value_counts().index)
plt.title('Crimes by Area')
plt.show()
```

```python
# Select numeric columns for correlation analysis
numeric_data = df.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
df.hist(figsize=(12, 10), bins=30)
plt.suptitle('Histograms of Numerical Features in Crime Data')
plt.show()
```

```python
# Get the top 10 most frequent crime types
top_crimes = df['Crm Cd Desc'].value_counts().nlargest(10).index

# Filter the data for these top crimes
filtered_data = df[df['Crm Cd Desc'].isin(top_crimes)]

# Create the box plot
plt.figure(figsize=(20,15))
sns.boxplot(x='Crm Cd Desc', y='Vict Age', data=filtered_data)
plt.title('Victim\'s Age by Top 10 Crime Types')
plt.xlabel('Crime Type')
plt.ylabel('Victim\'s Age')
plt.xticks(rotation=403)  # Rotate x labels for better readability
plt.show()
```
