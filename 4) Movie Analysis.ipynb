```python
# Distribution of movie ratings for data1
plt.figure(figsize=(10, 6))
sns.histplot(df['IMDb Rating'], bins=10, kde=True)
plt.title('Distribution of Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Rating distribution by release year for data1
plt.figure(figsize=(10, 6))
sns.lineplot(x=df['Year'], y='IMDb Rating', data=df)
plt.title('Rating Distribution by Release Year')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.show()
```
