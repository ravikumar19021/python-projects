```python
# Analyzing the data
#Day
plt.figure(figsize=(10, 8))
sns.histplot(df['cnt'], bins=30, kde=True)
plt.title('Distribution of Daily Rental Counts')
plt.xlabel('Rental Count')
plt.ylabel('Frequency')
plt.show()
```

```python
# Analyzing the data
#hour
plt.figure(figsize=(10, 8))
sns.histplot(df1['cnt'], bins=30, kde=True)
plt.title('Distribution of hour Rental Counts')
plt.xlabel('Rental Count')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='weekday', y='workingday', data=df)
plt.title('Weekday vs Workingday')
plt.xlabel('Weekday')
plt.ylabel('Workingday')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='weekday', y='workingday', data=df1)
plt.title('Weekday vs Workingday')
plt.xlabel('Weekday')
plt.ylabel('Workingday')
plt.show()
```

```python
sns.pairplot(df[['registered', 'casual', 'atemp', 'temp']])
plt.suptitle('Pair Plot of Daily Metrics', y=1)
plt.show()
```

```python
sns.pairplot(df1[['registered', 'casual', 'atemp', 'temp']])
plt.suptitle('Pair Plot of hourly Metrics', y=1)
plt.show()
```

```python
#Visualizing the data:
df['dteday'] = pd.to_datetime(df['dteday'], errors='coerce')
df['weekday'] = df['dteday'].dt.day_name()  # Extract day of the week from date
day_counts = df['weekday'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(day_counts, labels=day_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set2'))
plt.title('Distribution of Rentals by Day of the Week (Daily Data)')
plt.show()
```

```python
df1['dteday'] = pd.to_datetime(df1['dteday'], errors='coerce')
df1['weekday'] = df1['dteday'].dt.day_name()  # Extract day of the week from date
day_counts = df1['weekday'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(day_counts, labels=day_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set2'))
plt.title('Distribution of Rentals by hourly of the Week (Daily Data)')
plt.show()
```
